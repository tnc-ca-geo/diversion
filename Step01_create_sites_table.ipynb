```python
import pandas as pd
import os.path
```

```python
year = '2019'
insample = os.path.join( os.getcwd(), 'samples\\sites.csv' )
indata = os.path.join( os.getcwd(), '..', year, 'input_data\\stations.xlsx' ) # created by hand from ArcGIS and other sources
outname = os.path.join( os.getcwd(), '..', year, 'sites.csv' )
```

```python
df = pd.read_csv(insample)
df
```

```python
# Modify the comment for siphon 7

new_comment = "Water Right ID: S018557; AG3000 Meter Serial Number: 032017001576; Telemetry ID: 001D0A8920278. " + df.at[0, 'Comments']
df.at[0, 'Comments'] = new_comment
df
```

```python
dfxl = pd.read_excel(open(indata,'rb'), sheet_name="sites")
dfxl
```

```python
# Modify the sample table.  In this case, merge the sample data with the input data
df = df.append(dfxl, ignore_index=True)
df
```

```python
# Save as a new text file
df.to_csv(outname, encoding='utf-8', index=False)
```
