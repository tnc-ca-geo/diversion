```python
# Script to generate max diversion rate and monthly total amounts for water rights reporting requirements
```

```python
import pandas as pd
import numpy as np
import os.path
from pathlib import Path
from datetime import datetime
```

```python
in_methods = os.path.join( os.getcwd(), '..', 'methods.csv' )
in_sites = os.path.join( os.getcwd(), '..', 'sites.csv' )
in_sources = os.path.join( os.getcwd(), '..', 'sources.csv' )
in_Qualifiers = os.path.join( os.getcwd(), '..', 'Qualifiers.csv' )
in_variables = os.path.join( os.getcwd(), '..', 'variables.csv' )
in_filled = os.path.join( os.getcwd(), '..', 'DataValues_Reviewed_Filled_HourlyMean.csv' )

out_all = os.path.join( os.getcwd(), '..', 'Staten_Island_Water_Use_2018_CUAHSI_flat.csv' )
```

```python
#Import data
methods = pd.read_csv(in_methods)
print(list(methods))
sites = pd.read_csv(in_sites)
print(list(sites))
sources = pd.read_csv(in_sources)
print(list(sources))
Qualifiers = pd.read_csv(in_Qualifiers)
print(list(Qualifiers))
variables = pd.read_csv(in_variables)
print(list(variables))
df = pd.read_csv(in_filled)
print(list(df))
df
```

```python
# Create flat data frame 
df = pd.merge(df, sites, on='SiteCode', how='left')
df = pd.merge(df, variables, on='VariableCode', how='left')
df = pd.merge(df, methods, on='MethodCode', how='left')
df = pd.merge(df, Qualifiers, on='QualifierCode', how='left')
df
```

```python
#Remove extra fields
field_list = ['DateTimeUTC', 
              'SiteCode', 
              'MethodCode', 
              'UTCOffset', 
              'QualityControlLevelCode', 
              'DataValue', 
              'QualifierCode', 
              'LocalDateTime', 
              'SourceCode', 
              'CensorCode', 
              'VariableCode',
              'SiteName', 
              'Latitude', 
              'Longitude', 
              'LatLongDatumSRSName',
              'Comments', 
              'VariableName', 
              'Speciation', 
              'VariableUnitsName', 
              'SampleMedium', 
              'ValueType', 
              'IsRegular', 
              'TimeSupport', 
              'TimeUnitsName', 
              'DataType', 
              'GeneralCategory', 
              'NoDataValue',
              'MethodDescription', 
              'MethodLink',
              'QualifierDescription'
             ]

df = df[field_list]
list(df)
```

```python
print(df.MethodCode.unique().tolist())
print(df.SiteCode.unique().tolist())
```

```python
# Export data files to excel
methods = df.MethodCode.unique().tolist()
sites = df.SiteCode.unique().tolist()


for site in sites:
    if site == "Siphon7":
        method = 'AG3000_EM'
    else:
        method = 'IS203_EM'
    df_site = df.loc[(df['SiteCode'] == site) & (df['MethodCode'] == method)]
    file_name = os.path.join( os.getcwd(), '..', 'Staten_Island_Water_Use_' + site + '_'+ method +'_2018_CUAHSI_flat.csv' )
    df_site.to_csv(file_name, encoding='utf-8', index=False)
```

```python
# Export the report
df.to_csv(out_all, encoding='utf-8',index=False)
```
