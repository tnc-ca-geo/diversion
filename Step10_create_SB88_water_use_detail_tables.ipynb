```python
# Script to generate max diversion rate and monthly total amounts for water rights reporting requirements
```

```python
import pandas as pd
import numpy as np
import os.path
from pathlib import Path
from datetime import datetime
```

```python
year = '2019'
in_methods = os.path.join( os.getcwd(), '..', year, 'methods.csv' )
in_sites = os.path.join( os.getcwd(), '..', year, 'sites.csv' )
in_sources = os.path.join( os.getcwd(), '..', year, 'sources.csv' )
in_Qualifiers = os.path.join( os.getcwd(), '..', year, 'Qualifiers.csv' )
in_variables = os.path.join( os.getcwd(), '..', year, 'variables.csv' )
in_filled = os.path.join( os.getcwd(), '..', year, 'DataValues_Reviewed_Filled_HourlyMean.csv' )

out_all = os.path.join( os.getcwd(), '..', year, 'Staten_Island_Water_Use_{}_CUAHSI_flat.csv'.format(year) )
out_all_month = os.path.join( os.getcwd(), '..', year, 'Staten_Island_Water_Use_{}_Monthly.csv'.format(year) )
out_all_annual = os.path.join( os.getcwd(), '..', year, 'Staten_Island_Water_Use_{}_Annual.csv'.format(year) )
```

```python
#Import data
methods = pd.read_csv(in_methods)
print(list(methods))
sites = pd.read_csv(in_sites)
print(list(sites))
sources = pd.read_csv(in_sources)
print(list(sources))
Qualifiers = pd.read_csv(in_Qualifiers)
print(list(Qualifiers))
variables = pd.read_csv(in_variables)
print(list(variables))
df = pd.read_csv(in_filled)
print(list(df))
df
```

```python
# Create flat data frame 
df = pd.merge(df, sites, on='SiteCode', how='left')
df = pd.merge(df, variables, on='VariableCode', how='left')
df = pd.merge(df, methods, on='MethodCode', how='left')
df = pd.merge(df, Qualifiers, on='QualifierCode', how='left')
df
```

```python
#Remove extra fields
field_list = ['DateTimeUTC', 
              'SiteCode', 
              'MethodCode', 
              'UTCOffset', 
              'QualityControlLevelCode', 
              'DataValue', 
              'QualifierCode', 
              'LocalDateTime', 
              'SourceCode', 
              'CensorCode', 
              'VariableCode',
              'SiteName', 
              'Latitude', 
              'Longitude', 
              'LatLongDatumSRSName',
              'Comments', 
              'VariableName', 
              'Speciation', 
              'VariableUnitsName', 
              'SampleMedium', 
              'ValueType', 
              'IsRegular', 
              'TimeSupport', 
              'TimeUnitsName', 
              'DataType', 
              'GeneralCategory', 
              'NoDataValue',
              'MethodDescription', 
              'MethodLink',
              'QualifierDescription'
             ]

df = df[field_list]
list(df)
```

```python
print(df.MethodCode.unique().tolist())
print(df.SiteCode.unique().tolist())
```

```python
# Export data files to excel
methods = df.MethodCode.unique().tolist()
sites = df.SiteCode.unique().tolist()


for site in sites:
    if site == "Siphon7":
        method = 'AG3000_DL'
    else:
        method = 'IS203_EM'
    df_site = df.loc[(df['SiteCode'] == site) & (df['MethodCode'] == method)]
    file_name = os.path.join( os.getcwd(), '..', year, 'Staten_Island_Water_Use_' + site + '_'+ method +'_' + year + '_CUAHSI_flat.csv' )
    df_site.to_csv(file_name, encoding='utf-8', index=False)
```

```python
# Export the report
df.to_csv(out_all, encoding='utf-8',index=False)
```

## Create Monthly Summaries

```python
# Convert index to date index
df['LocalDateTime'] = pd.to_datetime(df['LocalDateTime'])
df.index = df['LocalDateTime']
del df['LocalDateTime']
df.head()
```

```python
df['Gallons'] = df.DataValue * 60
df_month = df.groupby(['SiteCode']).resample('M').sum().reset_index()
df_month
```

```python
df_month['AcreFeet'] = df_month.Gallons / 325851
df_month
```

```python
#Export monthly data
df_month[['SiteCode','LocalDateTime','AcreFeet']].to_csv(out_all_month, encoding='utf-8',index=False)
```

```python
#Export annual data
df_year = (df_month.groupby('SiteCode')
           .agg({'AcreFeet':'sum'})
           .reset_index()
          )
df_year
```

```python
df_year.to_csv(out_all_annual, encoding='utf-8',index=False)
```
