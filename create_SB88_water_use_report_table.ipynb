```python
# Script to generate max diversion rate and monthly total amounts for water rights reporting requirements
```

```python
import pandas as pd
import numpy as np
import os.path
from pathlib import Path
from datetime import datetime
```

```python
indata = os.path.join( os.getcwd(), '..', 'DataValues_Reviewed.csv' )
indata_filled = os.path.join( os.getcwd(), '..', 'DataValues_Reviewed_Filled_HourlyMean.csv' )
outdata_report = os.path.join( os.getcwd(), '..', 'SB88_WaterRightsReport.csv' )
```

```python
#Import raw datda
df = pd.read_csv(indata)
df.head()
```

```python
# Convert index to date index
df['LocalDateTime'] = pd.to_datetime(df['LocalDateTime'])
df.index = df['LocalDateTime']
#del df['LocalDateTime']
df
```

# Create Maximum Rate of Diversion from raw 5 min data (without outliers)

```python
# Set outlier values to missing

df.loc[df['QualifierCode'] == 1, 'DataValue'] = np.nan
```

```python

# Clip any 2019 data
mask = (df['LocalDateTime'] >= '2018-1-1') & (df['LocalDateTime'] < '2019-1-1')
df = df.loc[mask]
```

```python
# Generate the maximum rate for the DataValue column for each month
df_max = df.groupby(['SiteCode','MethodCode']).resample('M').max()
df_max = df_max["DataValue"].reset_index()
df_max
```

```python
# Convert the LocalDateTime to year and month
df_max["Month"] = df_max["LocalDateTime"].dt.month_name()
df_max["Year"] = df_max["LocalDateTime"].dt.year
df_max
```

```python
#Clean up columns
df_max.rename(columns={'DataValue': 'Maximum Rate of Diversion (GPM)', 'SiteCode':'Siphon','MethodCode':'Meter'}, inplace=True)
df_max.drop(['LocalDateTime'], axis=1, inplace=True)
df_max.head()
```

# Generate Monthly totals from hourly filled data

```python
# Read in the reviewed data with gaps filled
df = pd.read_csv(indata_filled)
df.head()
```

```python
# Convert index to date index
df['LocalDateTime'] = pd.to_datetime(df['LocalDateTime'])
df.index = df['LocalDateTime']
del df['LocalDateTime']
df.head()
```

```python
# Create monthly totals of amount diverted
df["Diverted (Gallons)"] = df['DataValue'] * 60 #Convert gallons per minue to gallons per hour
df_sum = df.groupby(['SiteCode','MethodCode']).resample('M').sum()
df_sum = df_sum["Diverted (Gallons)"].reset_index()
df_sum['Diverted (Acre Feet)'] = df_sum["Diverted (Gallons)"] / 325851
df_sum
```

```python
# Convert the LocalDateTime to year and month
df_sum["Month"] = df_sum["LocalDateTime"].dt.month_name()
df_sum["Year"] = df_sum["LocalDateTime"].dt.year
df_sum
```

```python
#Clean up columns
df_sum.rename(columns={'Diverted (Acre Feet)': 'Amount Directly Diverted (Acre Feet)', 'SiteCode':'Siphon','MethodCode':'Meter'}, inplace=True)
df_sum.drop(['LocalDateTime', 'Diverted (Gallons)'], axis=1, inplace=True)
df_sum.head()
```

```python
df_sum
```

```python
# Combine the two tables
match_fields = ['Siphon',
 'Meter',
 'Month',
 'Year']

df = pd.merge(df_sum, df_max, on=match_fields, how='outer')
```

```python
cols = df.columns.tolist()
```

```python
#Update the column sorting
cols =['Siphon',
       'Meter',
       'Year',
       'Month',
       'Maximum Rate of Diversion (GPM)',
       'Amount Directly Diverted (Acre Feet)']
df = df[cols]
df = df.fillna(0) 
df
```

```python
# Export the report
df.to_csv(outdata_report, encoding='utf-8',index=False)
```
